```python
from bs4 import BeautifulSoup
import requests
from csv import writer
import pandas as pd
from datetime import datetime
import json

start_time = datetime.now()

csv_file=pd.read_csv(r'Amazon Scraping - Sheet1.csv',usecols = ['Asin','country'])
Asin=csv_file['Asin'].tolist()
country=csv_file['country'].tolist()
Count=0

headers={ 
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9", 
    "Accept-Encoding": "gzip, deflate, br",
      
    "Upgrade-Insecure-Requests": "1", 
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36"}



for i, j in zip(Asin, country):
    url="https://www.amazon."+str(j)+"/dp/"+str(i)
    Count+=1
    page=requests.get(url,headers=headers)
    if page.status_code == 404 :
        print("url_not_available")
    
    if page.status_code == 200 and Count<5:
            soup = BeautifulSoup(page.content, 'html.parser')
            Product_Title=soup.find('span',class_="a-size-large product-title-word-break").text
            image_elements=soup.find('div',class_="imgTagWrapper")
            image_url=image_elements.find('img')['src']
            product_price=soup.find('span',class_="a-price-whole").text
            product_details=soup.find('ul',class_="a-unordered-list a-vertical a-spacing-mini").text
            
            data={"product_Title":Product_Title,"image_url":image_url,"product_price":product_price,"product_details":product_details }
    
            json_object=json.dumps(data)
            with open("sample.json", "a+") as outfile:
                outfile.write(json_object)
    
            
    
    if page.status_code==200 and Count>=5:
        try:
            soup = BeautifulSoup(page.content, 'html.parser')
            Product_Title=soup.find('span' ,class_="a-size-extra-large").text
            image_elements=soup.find('div',class_="maintain-height")
            image_url=image_elements.find('img')['src']
            product_price1=soup.find_all('span',class_="a-color-base")
            product_price=product_price1[1].text
            product_details=soup.find('div',class_="a-expander-content a-expander-partial-collapse-content").text
        
            data={
                "product_Title":Product_Title,
                "image_url":image_url,
                "product_price":product_price,
                "product_details":product_details 
                }
        
            json_object=json.dumps(data)
            with open("sample.json", "a+") as outfile:
                outfile.write(json_object)

        
        except AttributeError:
            print(url)
        
```
